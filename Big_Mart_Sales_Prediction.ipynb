```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.set_theme(color_codes=True)
```

```python
# import dataset
df=pd.read_csv('/Users/adithipai/Documents/GitHub/BigMartSalesPrediction/Train-Set.csv')
df.head()
```

```python
#Data Processing Part 1
#Check the number of unique values
df.select_dtypes(include='object').nunique()
```

```python
#Drop ProductID that is not unnecessary
df.drop(columns='ProductID', inplace=True)
df.head()
```

```python
#Exploratory Data Analysis
#List of categorical variable of plot
cat_vars = ['FatContent','ProductType','OutletID','OutletSize','LocationType','OutletType']

#create figure with subplots
fig, axs=plt.subplots(nrows=2, ncols=3, figsize=(20,10))
axs = axs.flatten()

#create barplot for each cateforical variable
for i, var in enumerate(cat_vars):
  sns.barplot(x=var, y='OutletSales',data=df, ax=axs[i])
  axs[i].set_xticklabels(axs[i].get_xticklabels(),rotation=90)

#adjust spacing between subplots
fig.tight_layout()

#show plot
plt.show()
```
